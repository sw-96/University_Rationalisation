# Passing Grade

```python
#Program untuk mengecek terpenuhinya passing grade
#Import library
import csv
import pandas as pd
import mpu.pd
import numpy as np

#Import data siswa
df_siswa = pd.read_csv('Data_Siswa_with_header.csv', delimiter = ',')
dic_siswa = df_siswa.to_dict('records')

Faktor_Konversi = 0.1 #Faktor konversi ke persentase


#Menambahkan key Persentase nilai ke data siswa
for i in np.arange(len(dic_siswa)):
    dic_siswa[i]['Persentase Nilai'] = dic_siswa[i]['Skor']*Faktor_Konversi


#Import data universitas
df_univ = pd.read_csv('Data_Univ_2_with_header.csv', delimiter = ',')
dic_univ = df_univ.to_dict('records')


#Membuat file csv untuk menampung hasil final
with open('Hasil_Final.csv', 'w', newline = '') as file:
    writer = csv.writer(file)

    writer.writerow(["Nama Siswa", "Skor", "Pilihan Universitas Pertama", "Pilihan Program Studi Pertama", "Status", "Pilihan Universitas Kedua", "Pilihan Program Studi Kedua", "Status"])

#Pengecekan pilihan universitas dan program studi
for i in np.arange(len(dic_siswa)):
    hasil = []
    for h in np.arange(2):
        if h == 0: #Pilihan 1
            for j in np.arange(len(dic_univ)):
                if dic_siswa[i]['Pilihan Universitas Pertama'] == dic_univ[j]['Universitas'] and dic_siswa[i]['Pilihan Program Studi Pertama'] == dic_univ[j]['Program Studi']:
                    #print(i,j,dic_siswa[i]['Nama Siswa'], dic_siswa[i]['Persentase Nilai'], dic_univ[j]['Universitas'], dic_univ[j]['Program Studi'], dic_univ[j]['Passing Grade (persen)'])
                    hasil.append(dic_siswa[i]['Nama Siswa'])
                    hasil.append(dic_siswa[i]['Persentase Nilai'])
                    hasil.append(dic_univ[j]['Universitas'])
                    hasil.append(dic_univ[j]['Program Studi'])
                    hasil.append(dic_univ[j]['Passing Grade (persen)'])

                    if dic_siswa[i]['Persentase Nilai'] >= dic_univ[j]['Passing Grade (persen)']:
                        #print('Memenuhi passing grade\n')
                        hasil.append('Memenuhi passing grade')
                    else:
                        #print('Tidak memenuhi passing grade\n')
                        hasil.append('Tidak memenuhi passing grade')
        else: #Pilihan 2
            for j in np.arange(len(dic_univ)):
                if dic_siswa[i]['Pilihan Universitas Kedua'] == dic_univ[j]['Universitas'] and dic_siswa[i]['Pilihan Program Studi Kedua'] == dic_univ[j]['Program Studi']:
                    #print(i,j,dic_siswa[i]['Nama Siswa'], dic_siswa[i]['Persentase Nilai'], dic_univ[j]['Universitas'], dic_univ[j]['Program Studi'], dic_univ[j]['Passing Grade (persen)'])
                    hasil.append(dic_univ[j]['Universitas'])
                    hasil.append(dic_univ[j]['Program Studi'])
                    hasil.append(dic_univ[j]['Passing Grade (persen)'])

                    if dic_siswa[i]['Persentase Nilai'] >= dic_univ[j]['Passing Grade (persen)']:
                        #print('Memenuhi passing grade\n')
                        hasil.append('Memenuhi passing grade')
                    else:
                        #print('Tidak memenuhi passing grade\n')
                        hasil.append('Tidak memenuhi passing grade')
    #print(hasil)
    with open('Hasil_Final.csv', 'a', newline = '') as file:
        writer = csv.writer(file)
        writer.writerow(hasil)

file.close()
```

# Rerata

```python
#Program untuk menghitung persentase perbedaan skor tes dengan skor rerata
#Import library
import csv
import pandas as pd
import mpu.pd
import numpy as np

#Import data siswa
df_siswa = pd.read_csv('Data_Siswa_with_header.csv', delimiter = ',')
dic_siswa = df_siswa.to_dict('records')

#Faktor_Konversi = 0.1 #Faktor konversi ke persentase


#Menambahkan key Persentase nilai ke data siswa
#for i in np.arange(len(dic_siswa)):
#    dic_siswa[i]['Persentase Nilai'] = dic_siswa[i]['Skor']*Faktor_Konversi


#Import data universitas
df_univ = pd.read_csv('Data_Univ_2_with_header.csv', delimiter = ',')
dic_univ = df_univ.to_dict('records')


#Menambahkan key nilai (bukan persentase) ke data universitas
for i in np.arange(len(dic_univ)):
    dic_univ[i]['Nilai'] = dic_univ[i]['Passing Grade (persen)']*10


#Membuat file csv untuk menampung hasil final
with open('Hasil_Final.csv', 'w', newline = '') as file:
    writer = csv.writer(file)

    writer.writerow(["Nama Siswa", "Skor", "Pilihan Universitas Pertama", "Pilihan Program Studi Pertama", "Rerata Pilihan Pertama", "Persentase 1", "Pilihan Universitas Kedua", "Pilihan Program Studi Kedua", "Rerata Pilihan Kedua", "Persentase 2"])

#Pengecekan pilihan universitas dan program studi
for i in np.arange(len(dic_siswa)):
    hasil = []
    for h in np.arange(2):
        if h == 0: #Pilihan 1
            for j in np.arange(len(dic_univ)):
                if dic_siswa[i]['Pilihan Universitas Pertama'] == dic_univ[j]['Universitas'] and dic_siswa[i]['Pilihan Program Studi Pertama'] == dic_univ[j]['Program Studi']:
                    #print(i,j,dic_siswa[i]['Nama Siswa'], dic_siswa[i]['Persentase Nilai'], dic_univ[j]['Universitas'], dic_univ[j]['Program Studi'], dic_univ[j]['Passing Grade (persen)'])
                    delta_nilai = round((dic_siswa[i]['Skor'] - dic_univ[j]['Nilai'])*100/dic_univ[j]['Nilai'],2)
                    hasil.append(dic_siswa[i]['Nama Siswa'])
                    hasil.append(round(dic_siswa[i]['Skor'],2))
                    hasil.append(dic_univ[j]['Universitas'])
                    hasil.append(dic_univ[j]['Program Studi'])
                    hasil.append(round(dic_univ[j]['Nilai'],2))
                    
                    if delta_nilai < 0:
                        text1 = "Lebih kecil " + str(abs(delta_nilai)) + "%" + " dari nilai rerata"
                        hasil.append(text1)
                    elif delta_nilai > 0:
                        text2 = "Lebih besar " + str(delta_nilai) + "%" + " dari nilai rerata"
                        hasil.append(text2)
                    else:
                        text3 = "Sama dengan nilai rerata"
                        hasil.append(text3)

                    #hasil.append(delta_nilai)


        else: #Pilihan 2
            for j in np.arange(len(dic_univ)):
                if dic_siswa[i]['Pilihan Universitas Kedua'] == dic_univ[j]['Universitas'] and dic_siswa[i]['Pilihan Program Studi Kedua'] == dic_univ[j]['Program Studi']:
                    #print(i,j,dic_siswa[i]['Nama Siswa'], dic_siswa[i]['Persentase Nilai'], dic_univ[j]['Universitas'], dic_univ[j]['Program Studi'], dic_univ[j]['Passing Grade (persen)'])
                    delta_nilai = round((dic_siswa[i]['Skor'] - dic_univ[j]['Nilai'])*100/dic_univ[j]['Nilai'],2)
                    hasil.append(dic_univ[j]['Universitas'])
                    hasil.append(dic_univ[j]['Program Studi'])
                    hasil.append(round(dic_univ[j]['Nilai'],2))

                    if delta_nilai < 0:
                        text1 = "Lebih kecil " + str(abs(delta_nilai)) + "%" + " dari nilai rerata"
                        hasil.append(text1)
                    elif delta_nilai > 0:
                        text2 = "Lebih besar " + str(delta_nilai) + "%" + " dari nilai rerata"
                        hasil.append(text2)
                    else:
                        text3 = "Sama dengan nilai rerata"
                        hasil.append(text3)
                    #hasil.append(delta_nilai)


    #print(hasil)
    with open('Hasil_Final.csv', 'a', newline = '') as file:
        writer = csv.writer(file)
        writer.writerow(hasil)

file.close()
```
